Read the code in from the text file and split each line into a list, and have each number in each line be an element of that created list 

```python
text_file_path = '/Users/rcomp/Documents/Code_projects/advent_of_code_2024/data/exercise_2_data.txt'
first_col_dict = {}
second_col_dict = {}

with open('/Users/rcomp/Documents/Code_projects/advent_of_code_2024/data/exercise_2_data.txt', 'r') as file: 
    for line in file: 
        line_split_list = line.split()
        line_split_list = [int(element) for element in line_split_list]
        # ensures the key exists with a default value of 0 if not present. Then + 1 increments the value.
        if line_split_list[0] in first_col_dict:
            first_col_dict[line_split_list[0]] += 1
        else:
            first_col_dict.setdefault(line_split_list[0], 1)
        if line_split_list[1] in second_col_dict:
            second_col_dict[line_split_list[1]] += 1
        else:
            second_col_dict.setdefault(line_split_list[1], 1)


        # first_col_dict[line_split_list[0]] = first_col_dict.setdefault(line_split_list[0], 1) + 1
        # second_col_dict[line_split_list[1]] = second_col_dict.setdefault(line_split_list[1], 1) + 1           
```

```python
print(first_col_dict)
```

```python
similarity_sum = 0

for key in first_col_dict:
    product = key * first_col_dict[key]
    if key in second_col_dict:
        product *= second_col_dict[key]
    elif key not in second_col_dict:
        product = 0
    similarity_sum += product
```
